```python
import re

en_file = open('../data/valid.en-zh.zh.sgm')
pattern = re.compile('<seg id=".*?"> (.*?) </seg>')
result=[]
for line in en_file.readlines():
    #print(line)
    item = re.findall(pattern,line)
    if item:
        print(item[0])
        result.append(item[0])
```

```python
print(len(result))
```
